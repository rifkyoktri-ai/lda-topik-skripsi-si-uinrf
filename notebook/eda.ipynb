# Exploratory Data Analysis (EDA)
Notebook ini digunakan untuk melakukan Explorasi Data Analysis (EDA) pada dataset skripsi.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Load Dataset
Dataset yang digunakan berada di folder `data/raw/`.

```python
# Load dataset
file_path = '../data/raw/dataset_skripsi_SI.csv'
df = pd.read_csv(file_path)

# Menampilkan 5 baris pertama
df.head()
```

## 2. Informasi Dataset
Melihat struktur data, tipe data, dan missing values.

```python
# Menampilkan info dataset
df.info()
```

```python
# Cek missing values
print("Missing Values per Kolom:")
df.isnull().sum()
```

```python
# Cek duplikat data
print('Jumlah duplikat:', df.duplicated().sum())
```

## 3. Analisis Univariat
### Distribusi Skripsi per Tahun

```python
# Countplot untuk Tahun
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Tahun', palette='viridis')
plt.title('Distribusi Jumlah Skripsi per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.show()
```

### Analisis Panjang Teks (Judul dan Abstrak)
Menambahkan kolom untuk melihat distribusi jumlah kata pada Judul dan Abstrak.

```python
# Menghitung jumlah kata
df['word_count_judul'] = df['Judul'].apply(lambda x: len(str(x).split()))
df['word_count_abstrak'] = df['Abstrak'].apply(lambda x: len(str(x).split()))

# Statistik deskriptif panjang teks
df[['word_count_judul', 'word_count_abstrak']].describe()
```

```python
# Distribusi panjang judul
plt.figure(figsize=(10, 6))
sns.histplot(df['word_count_judul'], bins=20, kde=True, color='blue')
plt.title('Distribusi Panjang Kata pada Judul Skripsi')
plt.xlabel('Jumlah Kata')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Distribusi panjang abstrak
plt.figure(figsize=(10, 6))
sns.histplot(df['word_count_abstrak'], bins=30, kde=True, color='green')
plt.title('Distribusi Panjang Kata pada Abstrak Skripsi')
plt.xlabel('Jumlah Kata')
plt.ylabel('Frekuensi')
plt.show()
```
